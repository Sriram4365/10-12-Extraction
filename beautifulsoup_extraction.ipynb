```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.collegenp.com/2-science-colleges/"

headers = {"X-Requested-With": "XMLHttpRequest"}
data = {"state": "on", "action": "filter", "count": "0"}

all_data = []
for page in range(0, 5):  # <-- increase number of pages here
    print("Getting page {}..".format(page))

    data["count"] = page * 10
    soup = BeautifulSoup(
        requests.post(url, data=data, headers=headers).content,
        "html.parser",
    )

    for c in soup.select(".college-name"):
        college_name = c.get_text(strip=True)
        location = c.find_next(class_="college-address").get_text(strip=True)


        all_data.append(
            {
                "College name": college_name,
                "Location": location,

            }
        )

df = pd.DataFrame(all_data)
print(df)
df.to_csv("data.csv", index=False)
```
